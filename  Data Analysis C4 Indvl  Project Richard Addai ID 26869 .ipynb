```python
# TASK 4 
import pandas as pd
netflix_df = pd.read_csv('netflix_data.csv')           # Create DataFrame netflix_df
netflix_df
```

```python
# TASK 4 CONTINUED
netflix_df.head()               # Print first five rows of netflix_df DataFrame for inspection
```

```python
# TASK 5
netflix_df_movies_only = netflix_df.loc[netflix_df['type'] == 'Movie']          # Subsetting netflix_df DataFrame to create netflix_df_movies_only DataFrame
netflix_df_movies_only
```

```python
# TASK 5 CONTINUED
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]            #Subset netflix_df_movies_only DataFrame to preserve specified columns and assign to new DataFrame netflix_movies_col_subset
netflix_movies_col_subset
```

```python
# TASK 5 continued

netflix_movies_col_subset.head()           # Print first 5 rows of netflix_movies_col_subset
```

```python
# TASK 6
import matplotlib.pyplot as plt
netflix_movies_col_subset.plot.scatter(x='release_year',y='duration',s=20)        # Create scatter plot for netflix_movies_col_subset DataFrame
plt.title('Movie Duration by Year of Release')                                    # Add title to the plot 
plt.show()                                                                        # show the plot
```

```python
# TASK 7
short_movies = netflix_movies_col_subset.loc[netflix_movies_col_subset['duration'] <= 60]                # Subsetting netflix_movies_col_subset dataframe to form short_movies dataframe
short_movies
```

```python
# TASK 7 continued
short_movies.head(20)                    # Print first 20 rows of short_movies DataFrame
```

```python
# TASK 8
colors = []                                                    # Initialize empty list called colors

netflix_movies_col_subset                                      #Using for loop to iterate through the netflix_movies_col_subset DataFrame's rows to append colors to list
for index, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')
colors                                                       # Print colors list
```

```python
# TASK 8 continued
print(colors[0:10])                      # Print first 10 values of colors list
```

```python
# TASK 9
netflix_movies_col_subset.plot.scatter(x='release_year',y='duration',s=20,c=colors)       # Creating scatter plot with color modifications using netflix_movies_col_subset dataframe 
plt.title('Movie Duration by Year of Release')                                            # Add title to the plot
plt.show()                                                                                # Show the plot
```
